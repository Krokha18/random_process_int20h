# Комбінуючи підходи, можна будувати бінарний класифікатор юзерів, що схильні відмінити підписку або ні.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json

from tqdm import tqdm
tqdm.pandas()
```

```python
df = pd.read_csv("int20h-ds-test-dataset.csv")
df['event_created_date'] = df['event_created_date'].apply(lambda x: pd.Timestamp(x))
df['event_attributes'] = df['event_attributes'].apply(lambda x: json.loads(x))

df.drop(columns = ['event_attributes', "device_model"], inplace = True)
df['event_name'] = df['event_name'].apply(lambda x: x.replace(" ",""))

df.head(5)
```

```python
users_df = df.groupby(by=['userid']).agg({i: list for i in df.columns.drop('userid')})
users_df
```

```python
def most_frequent(List):
    counter = 0
    if len(List)==0:
        return 'missing'
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
    return num
```

```python
users_df['user_state'] = users_df['user_state'].progress_apply(lambda x : 'missing' if x is None else most_frequent(x))
users_df['event_platform'] = users_df['event_platform'].progress_apply(lambda x : 'missing' if x is None else most_frequent(x))
users_df['device_manufacture'] = users_df['device_manufacture'].progress_apply(lambda x : 'missing' if x is None else most_frequent(x))
```

```python
users_df['events_list'] = users_df['event_name'].apply(lambda x: " ".join(x))
users_df.drop(columns=['event_name'], inplace=True)
```

```python
users_df['target'] = users_df['events_list'].apply(lambda x: "SubscriptionPremiumCancel" in x)
users_df['events_list'] = users_df['events_list'].apply(lambda x: x.replace("SubscriptionPremiumCancel", ""))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = lambda x: x.split(" "), ngram_range=(1,2), lowercase=False, max_features=50)

result = vectorizer.fit_transform(users_df['events_list'])

users_df[list(vectorizer.vocabulary_.keys())] = result.toarray()

users_df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

part1 = ohe.fit_transform(users_df[['user_state','event_platform','device_manufacture']])
part1_cols = list(np.hstack(ohe.categories_))

part1, part1_cols
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = lambda x: x.split(" "), ngram_range=(1,2), lowercase=False, max_features=50)

part2 = vectorizer.fit_transform(users_df['events_list'])
part2_cols = list(vectorizer.vocabulary_.keys())

part2, part2_cols
```

```python
#dates
#також можна провести роботу із датами

#наприклад, найбільш популярний день тижня, 
#найбільш популярний час користування, час між транзакціями (середній/максимальний)
```

```python
part1_cols
```

```python
X = pd.DataFrame(data = np.hstack([part1.toarray(), part2.toarray()]), columns = part1_cols+part2_cols)
y = users_df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify = y)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver='newton-cg', verbose=2)

log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import *
print(classification_report(y_test, y_pred))
```

```python
y_pred_proba = log_reg.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test, y_pred_proba>0.5))
```

```python
print(classification_report(y_test, y_pred_proba>0.9))
```

```python
print(classification_report(y_test, y_pred_proba>0.99))
```
