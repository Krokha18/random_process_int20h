```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

У цьому ноутбуці ми будемо аналізувати послідовність подій для кожного юзера <br>
- зробимо groupby і сформуємо events_list <br>
- складемо частотний опис (CountVectorizer - підхід як до аналізу тексту - пошук уніграм і біграм подій) <br>
- порівняємо розподіл і числові характеристики для юзерів що відмінили і не відмінили підписки <br>

```python
df = pd.read_csv("int20h-ds-test-dataset.csv", usecols = ['userid','event_name'])
df['event_name'] = df['event_name'].apply(lambda x: x.replace(" ",""))
df.head(5)
```

```python
events_list = df.groupby(by=['userid']).agg(list)
events_list['target'] = events_list['event_name'].apply(lambda x: "SubscriptionPremiumCancel" in x)
events_list['len'] = events_list['event_name'].apply(len)
events_list['events_list'] = events_list['event_name'].apply(lambda x: "|".join(x))
events_list.drop(columns=['event_name'], inplace=True)

events_list
```

Порівняємо частоти уніграм подій

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = lambda x: x.split("|"), ngram_range=(1,1), lowercase=False)

result = vectorizer.fit_transform(events_list['events_list'])

count_df = pd.DataFrame(data = result.toarray(), columns = list(vectorizer.vocabulary_.keys()), index = events_list.index)
count_df
```

Порахуємо частоти подій, порівняємо розподіл

```python
#users that have SubscriptionPremiumCancel event
top15_true = count_df[events_list['target']].mean().sort_values(ascending=False)[:15]
top15_true
```

```python
#users that don't have SubscriptionPremiumCancel event
top15_false = count_df[~events_list['target']].mean().sort_values(ascending=False)[:15]
top15_false
```

```python
top15_true.index.difference(top15_false.index)
```

```python
top15_false.index.difference(top15_true.index)
```

# З цього можна зробити невеличкий висновок - події AddVehicleFailed та SignUpError виникали частіше у людей що відмінили підписку.

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = lambda x: x.split("|"), ngram_range=(2,2), lowercase=False)

result = vectorizer.fit_transform(events_list['events_list'])

count_df = pd.DataFrame(data = result.toarray(), columns = list(vectorizer.vocabulary_.keys()), index = events_list.index)
count_df
```

```python
#users that have SubscriptionPremiumCancel event
top15_true = count_df[events_list['target']].mean().sort_values(ascending=False)[:15]
top15_true
```

```python
#users that don't have SubscriptionPremiumCancel event
top15_false = count_df[~events_list['target']].mean().sort_values(ascending=False)[:15]
top15_false
```

```python
top15_true.index.difference(top15_false.index)
```

```python
top15_false.index.difference(top15_true.index)
```

```python
top15_true.index.intersection(top15_false.index)
```
